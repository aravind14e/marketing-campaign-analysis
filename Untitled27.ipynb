```python
import pandas as pd

# Load data
df = pd.read_csv("/content/marketing_campaign_data.csv")

# Preview
df.head()

# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Feature engineering (example: Age groups)
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 25, 35, 45, 55, 65], labels=['18-25', '26-35', '36-45', '46-55', '56-65'])

# Save cleaned version
df.to_csv("/content/marketing_campaign_data.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/marketing_campaign_data.csv")

# Response rate by channel
sns.barplot(data=df, x='Channel', y='Responded')
plt.title("Response Rate by Channel")

# ROI distribution
sns.boxplot(data=df, x='Channel', y='ROI')
plt.title("ROI by Marketing Channel")

# Spend vs Revenue
sns.scatterplot(data=df, x='Marketing_Spend', y='Revenue', hue='Responded')
plt.title("Spend vs Revenue")
```

```python
# LTV = average revenue per responded user  response rate
responded = df[df['Responded'] == 1]
ltv = responded.groupby("Channel")['Revenue'].mean() * df.groupby("Channel")['Responded'].mean()

ltv = ltv.reset_index().rename(columns={0: "Estimated_LTV"})
print(ltv)

# Simple classification model (optional)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

features = df[['Age', 'Marketing_Spend', 'Cost']]
target = df['Responded']

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
model = LogisticRegression()
model.fit(X_train, y_train)
preds = model.predict(X_test)

print(classification_report(y_test, preds))
```

```python
# Total ROI per channel
roi_summary = df.groupby('Channel').agg({
    'Revenue': 'sum',
    'Cost': 'sum',
    'Marketing_Spend': 'sum'
}).reset_index()

roi_summary['ROI'] = ((roi_summary['Revenue'] - roi_summary['Cost']) / roi_summary['Cost']).round(2)
roi_summary.sort_values(by='ROI', ascending=False, inplace=True)

print(roi_summary)

# Pie chart of optimized budget
import matplotlib.pyplot as plt

optimal_allocation = roi_summary.copy()
optimal_allocation['Optimal_Share'] = optimal_allocation['ROI'] / optimal_allocation['ROI'].sum()

plt.pie(optimal_allocation['Optimal_Share'], labels=optimal_allocation['Channel'], autopct='%1.1f%%')
plt.title("Recommended Marketing Budget Allocation")
```
